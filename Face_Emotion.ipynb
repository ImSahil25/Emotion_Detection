```python
import cv2
from deepface import DeepFace
import matplotlib.pyplot as plt

# Load the image
img_path = '/Users/sahiladhikary/Desktop/Emotions/happy2.jpeg'
img = cv2.imread(img_path)

if img is None:
    raise ValueError("Image not loaded. Check the path.")

# Display the image
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()

# Try different backends
backends = ['opencv', 'ssd', 'mtcnn', 'dlib']

for backend in backends:
    try:
        predictions = DeepFace.analyze(img_path, detector_backend=backend, enforce_detection=False)
        print(f"Backend: {backend}")
        print(predictions)
        break  # If one backend works, no need to try others
    except Exception as e:
        print(f"Backend {backend} failed with error: {e}")
```
